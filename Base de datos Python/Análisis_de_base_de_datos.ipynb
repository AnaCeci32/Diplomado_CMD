```python
df = df.rename(columns = {'Date / Time':'Date/Time'})
df.head()
```

```python
df.groupby(['Date/Time']).City.nunique()
```

```python
df1 = pd.DataFrame(df.value_counts('City')).rename(columns = {0: 'Shape'}).reset_index()
df1
```

```python
df1 = pd.DataFrame(df.value_counts('City')).rename(columns = {0: 'Count'}).reset_index()
df1
    
```

```python
df2 = pd.DataFrame(df.groupby(['City']).Shape.value_counts()).rename(columns = {'Shape':'Count'}).reset_index()
df2
```

```python
fig = px.scatter(df2, x = 'City', y = 'Count', color = 'Shape')
fig
```
